```python
# -*- coding: utf-8 -*-
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import NoAlertPresentException
import unittest, time, re


driver = webdriver.Chrome()
driver.implicitly_wait(30)
base_url = "https://www.google.com.tw/"

driver.get(base_url + "/search?q=%E8%B5%B0%E9%A6%AC%E7%80%A8%E8%BE%B2%E5%A0%B4")
driver.find_element_by_id("_fZl").click()
```

```python
#driver.find_element_by_id("_fZl").click()
driver.find_element_by_css_selector('a[data-reply_source="local-search-reviews-list"]').click()
```

```python
eula = driver.find_element_by_css_selector('.review-dialog-list')
driver.execute_script('arguments[0].scrollIntoView(true);', eula)
driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', eula)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')
userreview = []
for reviews in soup.select('.gws-localreviews__google-review'):
    #print(reviews)
    username = reviews.select('._e8k')[0].text
    score = reviews.select('._pxg')[0]['aria-label']
    comments = reviews.select('._ucl')[0].text
    likes = reviews.select('._sEu')[0]['aria-label']
    userreview.append([username, score, comments, likes])
```

```python
import pandas
df = pandas.DataFrame(userreview)
df.columns = ['username', 'score', 'comment', 'likes']
df.to_csv('走馬瀨.csv')
```

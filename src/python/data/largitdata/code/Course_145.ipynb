```python
import requests
from datetime import datetime
import time
import re
res = requests.get('https://www.hkex.com.hk/Market-Data/Securities-Prices/Equities/Equities-Quote?sym=700&sc_lang=zh-HK')
m = re.search('LabCI.getToken = function \(\) { (.+?); };', ' '.join(res.text.split()) )
token = m.group(1).split('return')[-1].replace('"','').strip()
dt = int(time.mktime(datetime.now().timetuple()))
res = requests.get(f'https://www1.hkex.com.hk/hkexwidget/data/getchartdata2?hchart=1&span=0&int=0&ric=0700.HK&token={token}&qid={dt}&callback=jQuery_{dt}')
#res.text
```

```python
m2 = re.match('jQuery_\d+\((.+)\)',res.text)
```

```python
import json
jd = json.loads(m2.group(1))
```

```python
import pandas
df = pandas.DataFrame(jd['data']['datalist'])
```

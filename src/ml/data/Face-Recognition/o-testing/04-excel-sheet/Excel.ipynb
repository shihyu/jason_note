```python
import openpyxl
from datetime import datetime

# Load the Excel sheet
workbook = openpyxl.load_workbook('ABC.xlsx')
sheet = workbook.active

# Get the current date
current_date = datetime.today().strftime('%d-%m-%Y')

# List of student names to update
student_names = ["Sahil","Saurabh"]

# Find the column for the current date or add it if it doesn't exist
current_date_column = None
for col, header in enumerate(sheet[1]):
    if header.value == current_date:
        current_date_column = col + 1  # Add 1 to convert from 0-based index to 1-based index
        break
else:
    # If the current date column doesn't exist, create it
    current_date_column = len(sheet[1]) + 1
    sheet.cell(row=1, column=current_date_column, value=current_date)

# Iterate through the list of student names and update attendance for each student
for student_name in student_names:
    # Find the student's row index
    student_row_index = None
    for row_index, row in enumerate(sheet.iter_rows(min_row=2, max_col=1, values_only=True), start=2):
        if row[0] == student_name:
            student_row_index = row_index
            break

    if student_row_index is not None:
        # Update the attendance for the student on the current date
        sheet.cell(row=student_row_index, column=current_date_column, value='Present')
    else:
        print(f"Student '{student_name}' not found.")

# Save the updated Excel sheet
workbook.save('attendance_updated.xlsx')
```

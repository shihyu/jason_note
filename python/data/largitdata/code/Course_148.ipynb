```python
import asyncio, time
from pyppeteer import launch

async def main():
    browser = await launch(headless = False)
    page = await browser.newPage()
    await page.goto('https://24h.pchome.com.tw/prod/DYAJJU-1900BSGWO')
    time.sleep(1)
    page_text = await page.content()
    time.sleep(1)
    print(page_text)
    await browser.close()
    

asyncio.create_task(main())
```
